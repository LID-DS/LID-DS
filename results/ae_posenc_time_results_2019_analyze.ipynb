```python
import pandas as pd

# read the csv
df = pd.read_csv('ae_posenc_time_2019_results.csv')
df["CFA"] = df["consecutive_false_positives_normal"] + df["consecutive_false_positives_exploits"]
df = df.groupby(["stream_window"]).mean()
df["cost"] = df["CFA"] / df["detection_rate"] / df["exploit_count"]
df = df[["detection_rate","CFA","cost"]]
df.sort_values(by="cost",ascending=True)
```
